```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
g = np.linspace(-5,15,100)

def dgdt(g, s):
    return -g + 10*g*g/(16+g*g)+s
```

```python
fig = plt.figure(figsize=(6.4*1.2,4.8*1.2))
plt.plot(g, dgdt(g,0), label=r'Plot for $s=0$')
plt.plot(g, dgdt(g,3), label=r'Plot for $s=3$')

plt.title(r'Plot of $\frac{dg}{dt}$ vs $g$ for $\frac{dg}{dt}=-g+\frac{10g^2}{16+g^2}+s$')

plt.xlabel("Gene concentration ($g$)")
plt.ylabel(r'Gene concentration rate of change ($\frac{dg}{dt}$)')

plt.xlim([min(g), max(g)])

plt.axhline(y=0.0, color='r', linestyle='--')
plt.tick_params(direction='in')
plt.grid()
plt.savefig('/home/kiwi/Documents/Uni/math-260/a2/images/plot.eps')

plt.legend()

# plt.savefig('/home/kiwi/Documents/Uni/phy-121/latex/Lab 3/Resources/processed-small.eps')
plt.show()
```
