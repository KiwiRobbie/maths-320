```python
N = [
    2,
    3,
    4,
    5,
    7,
    8,
    9,
    11,
    13
]
```

```python
def proj_params(order):
    v = n**2 + n + 1 
    b = n**2 + n + 1 
    
    r = n + 1 
    k = n + 1 
    
    l = 1 
    return (v, b, r, k, l)

def affine_params(order):
    v = n**2 
    b = n**2 + n 
    
    r = n + 1 
    k = n 
    
    l = 1 
    return (v, b, r, k, l)

def k_subset_params(v, k):
    b = comb(v,k)
    r = comb(v-1, k-1)
    l = comb(v-2, k-2)
    return (
        v, b, r, k, l
    )

def complement_params(v, b, r, k, l):
    return (
        v,
        b,
        b-r,
        v-k,
        b-2*r+l
    )

```

```python
print("Projective plane (order n)")
print(f" n   v   b   r  k  l")
print(f"--------------------")
for n in N:
    v, b, r, k, l = proj_params(n)
    print(f"{n:2}: {v:3} {b:3} {r:2} {k:2}  {l}", )
```

```python
print("Affine plane (order n)")
print(f" n   v   b   r  k  l")
print(f"--------------------")
for n in N:
    v, b, r, k, l = affine_params(n)
    print(f"{n:2}: {v:3} {b:3} {r:2} {k:2}  {l}", )
```
